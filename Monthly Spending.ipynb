```python
cur_month = datetime.now().month
month_days = [monthrange(2020,cur_month)[1]]
print('Avg Spend per day: ${}'.format((spend/delta).values))
spnd_pr_day = (spend/delta).values
print('Left over budget: ${}'.format(round(left_over_spend,2)))
spnd = spnd_pr_day.tolist()
pace = month_days[0] * spnd[0]
print('Current Month spending Pace: ${}'.format(pace))
monthly = df.resample('M', closed='right').sum()
print('Pace to Budget Difference: ${}'.format(monthly.DailyBudget[0] - pace))
print('New Daily Budget: ${}'.format(monthly.Difference[0]/((month_days) - MTD)))
```
